```python
import pandas as pd
import os
import numpy as np
# Read the Excel file
df = pd.read_csv(r"C:\Users\AbhishekDubey\PycharmProjects\MJ-Pro\usernames.csv")

# Create a new DataFrame to store the transformed data
new_columns = ['City', 'Date', 'Amount', 'Percentage']
new_df = pd.DataFrame(columns=new_columns)

# Iterate through each row in the original DataFrame
for index, row in df.iterrows():
    city = str(row['City'])
    date = str(row['Date'])
    amount = row['Amount']
    percentage = row['Percentage']

    # Check the data type of each cell
    if pd.notna(amount) and isinstance(amount, (int, float)):
        new_df = new_df.append({'City': city, 'Date': date, 'Amount': amount, 'Percentage': np.nan}, ignore_index=True)
    elif pd.notna(percentage) and isinstance(percentage, (int, float)):
        new_df = new_df.append({'City': city, 'Date': date, 'Amount': np.nan, 'Percentage': percentage}, ignore_index=True)
    elif pd.notna(amount) and isinstance(amount, str) and '%' in amount:
        new_df = new_df.append({'City': city, 'Date': date, 'Amount': np.nan, 'Percentage': amount}, ignore_index=True)
    elif pd.notna(percentage) and isinstance(percentage, str) and '%' in percentage:
        new_df = new_df.append({'City': date, 'Date': amount, 'Amount': np.nan, 'Percentage': percentage}, ignore_index=True)
    else:
        new_df = new_df.append({'City': city, 'Date': date, 'Amount': amount, 'Percentage': percentage}, ignore_index=True)

# Write the new DataFrame to a new Excel file
output_file_path = os.path.join(r'C:\Users\AbhishekDubey\Desktop\Orient', 'output_excel_file.xlsx')
new_df.to_excel(output_file_path, index=False)

print(f"Output written to {output_file_path}")


```

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\AbhishekDubey\PycharmProjects\MJ-Pro\usernames-name.csv")

# Initialize lists for each column
cities, dates, amounts, percentages, names = [], [], [], [], []

# Iterate through rows
for index, row in df.iterrows():
    city, date, amount, percentage, name = '', '', None, None, ''
    
    # Iterate through items in the row
    for item in row:
        if item.isdigit():
            amount = int(item)
        elif '%' in item:
            percentage = item
        elif '/' in item or '-' in item:
            date = item
        else:
            # Check if the item contains alphabets, assuming it's a name
            if any(c.isalpha() for c in item):
                name = item
            else:
                city = item
    
    # Append values to lists
    cities.append(city)
    dates.append(date)
    amounts.append(amount)
    percentages.append(percentage)
    names.append(name)

# Create a new DataFrame
new_data = {'City': cities, 'Date': dates, 'Amount': amounts, 'Percentage': percentages, 'Name': names}
new_df = pd.DataFrame(new_data)

# Print the result
print(new_df)
```

```python
import pandas as pd

# Sample data
data = {'City': ['Gorakhpur', '09-10-2023', '01-01-2023', 'Gorakhpur'],
        'Date': ['08-10-2023', 'Pune', 'Delhi', '20-02-2021'],
        'Amount': ['2421', '67%', 'Lucknow', '1278'],
        'Percentage': ['', '100%', '1278', '70%']}
df = pd.DataFrame(data)

# Initialize lists for each column
cities, dates, amounts, percentages = [], [], [], []

# Iterate through rows
for index, row in df.iterrows():
    city, date, amount, percentage = '', '', '', ''
    
    # Iterate through items in the row
    for item in row:
        if item.isdigit():
            amount = int(item)
        elif '%' in item:
            percentage = item
        elif '/' in item or '-' in item:
            date = item
        elif item != '':
            city = item
    
    # Append values to lists
    cities.append(city if city else None)
    dates.append(date if date else None)
    amounts.append(amount if amount else None)
    percentages.append(percentage if percentage else None)

# Create a new DataFrame
new_data = {'City': cities, 'Date': dates, 'Amount': amounts, 'Percentage': percentages}
new_df = pd.DataFrame(new_data)

# Forward-fill missing values in the 'City' column
new_df['City'] = new_df['City'].fillna(method='ffill')

# Print the result
print(new_df)
```
